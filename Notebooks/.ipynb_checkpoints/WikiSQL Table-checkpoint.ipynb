```python
text_list = ["Largest stadiums in Europe \n\n\n"]

def make_text(row):
#     print(row)
    if row.position == 1:
        string = "The stadium with "+str(row.ordinal_position)+" capacity is "+str(row.Stadium)+". The capacity is "+str(row.Capacity)+". It is located in "+str(row.City)+", "+str(row.Country)+". The Home Team is "+str(row.Tenant)+". The stadium was opened in the year "+str(row.Opening)+"."
    else:
        string = "The "+str(row.ordinal_position)+" largest stadium is "+str(row.Stadium)+". The capacity is "+str(row.Capacity)+". It is located in "+str(row.City)+", "+str(row.Country)+". The Home Team is "+str(row.Tenant)+". The stadium was opened in the year "+str(row.Opening)+"."
#     print(string)
    text_list.append(string)
```

```python
stadium_df.apply(make_text,axis=1)
```
