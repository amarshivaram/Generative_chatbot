```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib.font_manager import FontProperties
import csv
```

```python
def readcsv(files):
    csvfile = open(files, 'r')
    plots = csv.reader(csvfile, delimiter=',')
    next(plots, None)
    x = []
    y = []
    for row in plots:
        x.append(float(row[1]))
        y.append(float(row[2])) 
    return x ,y
```

## Full run of 50 epochs graphs

```python

mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = 'NSimSun,Times New Roman'
 
 
plt.figure()
x2,y2=readcsv("run-r8_dropout01_final-tag-loss_train.csv")
plt.plot(x2, y2, color='red', label='Train')
 
x,y=readcsv("run-r8_dropout01_final-tag-loss_valid.csv")
plt.plot(x, y, 'g',label='Validation')

 
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.title("Blenderbot - dropout 0.1 - 50 epochs - loss", fontsize = 15) 
#plt.ylim(0, 12)
#plt.xlim(0, 45000)
plt.xlabel('Step',fontsize=12)
plt.ylabel('Value',fontsize=12)
plt.legend(fontsize=12)
plt.show()
```
